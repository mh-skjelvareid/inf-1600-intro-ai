```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Parameters (edit as needed)
total_samples = 500
samples_per_cycle_1 = 100
samples_per_cycle_2 = 250
amplitude_1 = -20
amplitude_2 = 20
offset = 90
noise_std = 7
random_seed = 42

noisy_signal_offset = 10
```

```python
# Time vector
t = np.arange(total_samples)

# Sine waves
sin1 = amplitude_1 * np.sin(2 * np.pi * t / samples_per_cycle_1)
sin2 = amplitude_2 * np.sin(2 * np.pi * t / samples_per_cycle_2)

# Clean signal
signal_clean = sin1 + sin2 + offset
```

```python
# Add noise using numpy Generator
gen = np.random.default_rng(random_seed)
noise = gen.normal(loc=0, scale=noise_std, size=total_samples)
signal_noisy = signal_clean + noise + noisy_signal_offset
```

```python
# Plot
plt.plot(t, signal_noisy)
plt.plot(t, signal_clean)
plt.show()
```

```python
# Save both clean and noisy time series as two columns in the same CSV file
out_df = pd.DataFrame(
    {"t": t, "heart_rate_accurate": signal_clean, "heart_rate_noisy": signal_noisy}
)
out_df = out_df.round(2)
out_df.to_csv("noisy_heart_rate_time_series.csv", index=False)
print("CSV file saved: noisy_sine_timeseries.csv")
```
